```python
#1. Import Library Yang di perlukan
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# muat data iris
iris = load_iris()
x = iris.data #Fitur (ukuran kelopak dan sepal)
y = iris.target #Label (spesies iris)D
```

```python
# 2.Memuat dataset iris
```

```python
print("Nama fitur:", iris.feature_names)
print("Nama Target (Spesies)", iris.target_names)
```

```python
print("Bentuk Data x :", x.shape)
print("Bentuk Data y :", y.shape)
```

```python
# Opsional: Tampilkan beberapa baris fitur
print("\n Beberapa baris pertama data fitur(x) :\n", x[:5])
print("\n Beberapa baris pertama label(y) :\n", y[:5])
```

```python
# 3.Membagi data menjadi pelatihan dan pengujian
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

print("\n Bentuk X_train:", x_train.shape)
print(" Bentuk X_test:", x_test.shape)
print(" Bentuk Y_train:", y_train.shape)
print(" Bentuk Y_train:", y_test.shape)
```

```python
# 4.Memilih dan melatih model supervised Learning (K-Nearest Neighbors)
```

```python
#Pilih model supervised learning (Contoh: K-Nearest Neighbors)
#n_neighbors adalah parameter K dalam KNN

knn = KNeighborsClassifier(n_neighbors=3)

print("\n Model KNN Telah dilatih.")
```

```python
#5. Membuat Prediksi
```

```python
# Buat prediksi pada data pengujian
# Latih model menggunakan data pelatihan
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)

#Tampilkan beberapa hasil prediksi dibandingkan aktual
print("\nPrediksi pada data pengujian (y_pred):\n", y_pred[:10])
print("Nilai aktual data pengujian (y_test):\n", y_test[:10])
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=iris.target_names,
            yticklabels=iris.target_names)
plt.xlabel("Prediksi")
plt.ylabel("Aktual")
plt.title("Confusion Matrix KNN pada Dataset Iris")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=x[:,0], y=x[:,1], hue=iris.target_names[y], palette="Set1", s=80, edgecolor="k")

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sebaran Data Iris berdasarkan Sepal Length & Sepal Width")
plt.legend(title="Spesies")
plt.show()
```
